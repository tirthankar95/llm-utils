```python
import torch
torch.cuda.empty_cache()

def get_vram_usage():
    """Get current VRAM usage in MB"""
    if torch.cuda.is_available():
        allocated = torch.cuda.memory_allocated() / 1024**2
        reserved = torch.cuda.memory_reserved() / 1024**2
        total = torch.cuda.get_device_properties(0).total_memory / 1024**2
        return allocated, reserved, total
    return 0, 0, 0

# Initial state
allocated, reserved, total = get_vram_usage()
print(f"Initial - Free: {(total-allocated):.2f} MB,  Reserved: {reserved:.2f} MB")

# Create tensor
x = torch.randn(10000, 1000).to('cuda')
allocated, reserved, total = get_vram_usage()
print(f"After - Free: {(total-allocated):.2f} MB, Reserved: {reserved:.2f} MB")
```

```python
del x 
'''reserved memory may not be freed immediately, it's used internally by PyTorch for performance'''
torch.cuda.empty_cache()
allocated, reserved, total = get_vram_usage()
print(f"Final - Free: {(total-allocated):.2f} MB, Reserved: {reserved:.2f} MB")
```
