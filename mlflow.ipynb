```python
import mlflow
import mlflow.sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
```

mlflow.set_experiment(): Organizes related runs. All runs without an explicit experiment name go to the Default experiment.

with mlflow.start_run():: Creates a new Run. All code and logging calls within this block are recorded to that specific run.

mlflow.log_param(key, value): Logs a single hyperparameter or configuration setting.

mlflow.log_metric(key, value): Logs a single evaluation metric. You can call this multiple times (e.g., inside a training loop) to track metrics over time.

mlflow.sklearn.log_model(model, artifact_path): Logs the trained model. artifact_path specifies where the model is stored within the run's artifact directory. The model is saved in the MLflow Model format.
